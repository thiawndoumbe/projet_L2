```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service as ChromeService

def automate_login(url, email, password, chromedriver_path):
    options = webdriver.ChromeOptions()
    # options.add_argument('--headless')
    options.add_argument('--no-sandbox')

    driver = webdriver.Chrome(options=options, service=ChromeService(chromedriver_path))
    
    driver.get(url)

    try:
        mail = WebDriverWait(driver, 60).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "input[name=loginfmt]"))
        )
        suivant = WebDriverWait(driver, 60).until(
            EC.presence_of_element_located((By.ID, "idSIButton9"))
        )

        mail.send_keys(email)
        suivant.click()
        time.sleep(2)
        
        pwd = WebDriverWait(driver, 60).until(
            EC.presence_of_element_located((By.ID,"i0118"))
        )
        connecter = WebDriverWait(driver, 60).until(
            EC.presence_of_element_located((By.ID, "idSIButton9"))
        )

        pwd.send_keys(password)
        connecter.click()
        
        element = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.ID, "idBtn_Back"))
        )
        element.click()

        time.sleep(2)
        s = WebDriverWait(driver, 60).until(
            EC.presence_of_element_located((By.CLASS_NAME, "profile-picture-container"))
        )
        time.sleep(2)
        s.click()

        time.sleep(2)
    except:
        print("Erreur lors de la connexion automatique")

    finally:
        driver.quit()

if __name__ == "__main__":
    URL = "https://teams.microsoft.com/"
    EMAIL = "Thiawndoumbe@mail.dit.sn"
    PASSWORD = "Qus28093"
    CHROMEDRIVER_PATH = "C:\\Users\\hp\\Desktop\\Projet_fin_etude\\Drivers\\.wdm\\drivers\\chromedriver\\win32\\114.0.5735.90\\chromedriver.exe"
    
    automate_login(URL, EMAIL, PASSWORD, CHROMEDRIVER_PATH)
```
